```python
import numpy as np
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.datasets import fetch_california_housing
np.random.seed(25)
```

```python
house = fetch_california_housing(as_frame=True)
y = house.target
x = house.data

```

```python
x
```

```python
def minmax(x):
    x_min = x.min()
    x_max = x.max()
    
    return (x - x.min(axis=0)) / (x.max(axis=0) - x.min(axis=0))

x_min = x.min().values
x_max = x.max().values
y_min = y.min()
y_max = y.max()
x_min
```

```python
x_scaled = minmax(x)
y_scaled  = minmax(y)
x_scaled
```

```python
x_train,x_test,y_train,y_test = train_test_split(x_scaled,y_scaled,test_size=0.2,random_state=42)


print(f"X_train: {x_train.shape} Y_train: {y_train.shape} X_test: {x_test.shape} Y_test: {y_test.shape}")
```

```python
epochs = 1000
alpha = 0.1
w = np.random.rand(x_train.shape[1])
b = 0.0
```

```python
def pred(x, w, b):
    return np.dot(x, w) + b
```

```python
def model_training(x_train,w,b):

    for i in range(epochs):
        y_pred = pred(x_train, w, b)

        error = y_pred - y_train  # shape (n,)

        dw = (2 / len(x_train)) * np.dot(x_train.T, error)   
        db = (2 / len(x_train)) * np.sum(error)

        w -= alpha * dw
        b -= alpha * db

        if i % 100 == 0:
            mse = np.mean(error**2)
            print(f"Epoch {i}: Error {mse:.4f}")
        elif(i==1000):
            mse = np.mean(error**2)
            print(f"Epoch {i}: Error {mse:.4f}")       
    return w,b 
            
w,b = model_training(x_train,w,b)
```
